### Line Charts and Bar Charts

```python
movies_yearly = movies.Year.value_counts().sort_index(ascending = False)
movies_yearly.index
```

```python
plt.figure(figsize = (15,5))
plt.plot(movies_yearly)
plt.xticks(movies_yearly.index)
plt.ylim(top = 400, bottom = 0)
plt.title("No. of Movies produced each year between 2001 and 20018")
plt.show()
```

```python
genre_tbl = movies['Genre'].value_counts()
genre_tbl
```

```python
plt.figure(figsize = (15,5))
plt.plot(genre_tbl)
plt.xticks(rotation=45)
plt.title("No of movies produced in each genre between 2001 and 2018")
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.bar(x = genre_tbl.index, height = genre_tbl.values)
plt.xticks(rotation = 90)
plt.title("No of movies produced in each genre between 2001 and 2018")
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x = genre_tbl.values, y = genre_tbl.index, color = 'lightblue')
plt.title("No of movies produced in each genre between 2001 and 2018")
plt.show()
```

### Pie Charts

```python
# what is the % of movies with rating of less than 6, between 6 and 7 and above 7
```

```python
movies['Ratings_cat'] = np.where(movies.Rating >=7,"High",np.where(movies.Rating >=6, 'Medium', 'Low'))
```

```python
movies.head()
```

```python
movies['Ratings_cat'].value_counts(normalize = True)
```

```python
plt.pie(movies['Ratings_cat'].value_counts(),
       labels = movies['Ratings_cat'].value_counts().index,
       autopct = '%1.2f%%',
       explode=[0, 0, 0.1])
plt.title("% of Movies with High, Medium, Low rating");
```

### Two-way Tables

```python
# how many movies received high, medium and low rating each year
```

```python
pd.crosstab(movies.Year, movies.Ratings_cat)
```

```python
pd.crosstab(movies.Year, movies.Ratings_cat, normalize = True)
```

```python
pd.crosstab(movies.Year, movies.Ratings_cat, normalize = 'index')
```

```python
pd.crosstab(movies.Ratings_cat,movies.Year, normalize = 'columns')
```

### Heat Maps

```python
plt.figure(figsize = (10,10))
sns.heatmap(pd.crosstab(movies.Year, movies.Ratings_cat, normalize = 'index'), annot = True)
plt.yticks(rotation = 0);
```
